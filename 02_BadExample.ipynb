```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter  

# Define a function that formats numbers as percentage strings.
def to_percent(y, position):
    """Convert number to percentage string."""
    return f"{int(y)}%"  

# Data for plotting
years = ['2009', '2019']  # Define the x-axis data (years).
percentage = [77, 65]  # Define the y-axis data (percentages).

plt.figure(figsize=(8, 5))  # Set the figure size for the plot.
bars = plt.bar(years, percentage, color='steelblue')  # Create a bar plot using the years and percentage data.

# Add title and labels
plt.title('Percentage of Americans that Identify as Christians', pad=20, fontname='sans-serif')  # Add a title to the plot.

# Set the formatter for the Y-axis as percentage
formatter = FuncFormatter(to_percent)  # Create a formatter object using the to_percent function.
plt.gca().yaxis.set_major_formatter(formatter)  # Apply the formatter to the y-axis.

# Set the limits for the Y-axis
plt.ylim(0, 100)  # Set the limits of the y-axis from 0 to 100.

# Adding the percentage above each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of each bar (percentage value).
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}%', ha='center', va='bottom')  # Add text above each bar.

# Plot a line graph over the bar graph to indicate change
plt.plot(years, percentage, marker='o', color='darkorange')

# Display the plot
plt.show()

# Display the plot
plt.show()
```
